```python
import pandas as pd
import numpy as np
import sqlite3
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
from scipy.stats import ttest_ind,pearsonr
from scipy.stats import pearsonr
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score
from sklearn.ensemble import RandomForestClassifier

warnings.filterwarnings('ignore')
conn = sqlite3.connect('airlines.db')
df = pd.read_sql_query('select * from airlines_df',conn)

df['satisfaction_binary'] = (df['satisfaction'] == 'satisfied').astype(int)

def countplot(df,column,hue=None):
    sns.countplot(data=df,x=column,palette='coolwarm',hue=hue)
    plt.title(f'{column} Distribution')
    plt.tight_layout()
    plt.show()

#Identify customer segments with high value but low satisfaction#
def high_value_low_satisfaction(df):
    high_distance_threshold = df['flight distance'].quantile(0.75)
    high_value_mask = (
        (df['flight distance'] >= high_distance_threshold) |
        (df['is_frequently_flyer'] == 1) |
        (df['is_business_traveler'] == 1)
    )
    low_satisfaction_mask = (
        (df['satisfaction'] == 'neutral or dissatisfied') |
        (df['inflight_score_avg'] < 3) |
        (df['ground_score_avg'] < 3)
    )
    high_value_low_satisfaction = df[high_value_mask  & low_satisfaction_mask]
    return high_value_low_satisfaction
    
def segment_summary_correlation(df):
    segment_summary = df.groupby(['type of travel', 'class']).agg({
        'flight distance': 'mean',
        'inflight_score_avg': 'mean',
        'ground_score_avg': 'mean',
        'id': 'count'
    }).rename(columns={'id': 'customer_count'}).reset_index()
    numerical_cols = segment_summary.select_dtypes(include=np.number).columns
    segment_summary['travel_binary'] = (segment_summary['type of travel'] == 'Business travel').astype(int)
    le = LabelEncoder()
    segment_summary['class_type'] = le.fit_transform(segment_summary['class'])
    numerical_cols = segment_summary.select_dtypes(include=np.number).columns
    plt.title('Inflight Features Correlation Map')
    sns.heatmap(segment_summary[numerical_cols].corr(),annot=True,cmap='crest')
    plt.tight_layout()
    plt.show()
    
# Service Quality & Satisfaction Drivers 
# Assess which in-flight services (wifi, food, seat comfort) most impact overall satisfaction

def Inflight_features_correlation(df):
    inflight_features = ['inflight wifi service', 'food and drink', 'seat comfort',
        'inflight entertainment', 'on-board service',
        'leg room service', 'inflight service', 'cleanliness']
    correlations = {
        feature: pearsonr(df[feature], df['satisfaction_binary'])[0] 
        for feature in inflight_features
    }
    correlation_df = pd.DataFrame.from_dict(correlations, orient='index', columns=['Correlation']).sort_values(by='Correlation', ascending=False)
    return correlation_df
    
def barplot(df,column):
    plt.figure(figsize=(8,5))
    sns.barplot(x=df[column],y=df.index,palette='coolwarm')
    plt.tight_layout()
    plt.show()
    

# Perform statistical tests (e.g., correlation, t-tests) to validate key drivers
# Provide recommendations for service improvements

def delay_satisfaction(df):
    df['delay_bucket'] = pd.cut(df['delay_total'],bins=[-1,0,10,30,60,120,300,1000],labels = ['0','1-10','11-30','31-60','61-120','121-300','300+'])
    thresholds = [0, 10, 20, 30, 60, 90, 120, 180]
    for t in thresholds:
        df[f'delayed_over_{t}'] = (df['delay_total'] > t).astype(int)
        rate = df[df[f'delayed_over_{t}'] == 1]['satisfaction_binary'].mean()
        print(f"Avg satisfaction when delay > {t} mins: {rate:.2f}")
    seg_travel = df.groupby(['type of travel', 'delay_bucket'])['satisfaction_binary'].mean().reset_index()
    plt.figure(figsize=(10, 6))
    sns.lineplot(data=seg_travel, x='delay_bucket', y='satisfaction_binary', hue='type of travel', marker='o')
    plt.title("Satisfaction by Delay and Travel Type")
    plt.ylabel("Satisfaction Rate")
    plt.xlabel("Delay Bucket")
    plt.grid(True)
    plt.show()
    # Group by class and delay bucket
    seg_class = df.groupby(['class', 'delay_bucket'])['satisfaction_binary'].mean().reset_index()
    plt.figure(figsize=(10, 6))
    sns.lineplot(data=seg_class, x='delay_bucket', y='satisfaction_binary', hue='class', marker='o')
    plt.title("Satisfaction by Delay and Class")
    plt.ylabel("Satisfaction Rate")
    plt.xlabel("Delay Bucket")
    plt.grid(True)
    plt.show()    

def ttest_results(df):
    inflight_features = ['inflight wifi service', 'food and drink', 'seat comfort',
        'inflight entertainment', 'on-board service',
        'leg room service', 'inflight service', 'cleanliness']
    satisfied = df[df['satisfaction_binary'] == 1]
    unsatisfied = df[df['satisfaction_binary'] == 0]

    # T-test for each inflight feature    
    ttest_results = {}
    for feature in inflight_features:
        t_stat, p_val = ttest_ind(satisfied[feature], unsatisfied[feature], equal_var=False)
        ttest_results[feature] = {"t-stat": t_stat, "p-value": p_val}

    ttest_df = pd.DataFrame(ttest_results).T.sort_values(by="p-value")
    return ttest_df
def recommendations(ttest_df,correlation_df):
    significant_features = ttest_df[ttest_df['p-value'] < 0.05].index.tolist()
    print("\n✅ Recommendation Summary:")
    for feature in significant_features:
        impact = correlation_df.loc[feature, 'Correlation']
        if impact > 0.3:
            print(f"- Improve '{feature}': Strong positive impact on satisfaction (corr={impact:.2f})")
        elif impact > 0.1:
            print(f"- Consider improving '{feature}': Moderate positive impact (corr={impact:.2f})")
        else:
            print(f"- '{feature}' is significant but has low correlation (corr={impact:.2f})")

# Predictive Modeling: Churn Risk via Dissatisfaction
# Build a model to predict dissatisfied passengers using demographic, travel, and service data
# Use this to simulate churn likelihood and identify at-risk groups

def randomForestModelBuilding(df):
    
    df = df.dropna()
    features = [
        'gender', 'age', 'type of travel', 'class',
        'flight distance', 'inflight_score_avg', 'ground_score_avg',
        'delay_total', 'is_frequently_flyer', 'is_business_traveler',
        'is_long_haul', 'vip_delay_penalty', 'comfort_gap'
    ]
    X = df[features]
    y = df['satisfaction_binary']
    X_encoded = pd.get_dummies(X,drop_first=True)
    X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)
    clf = RandomForestClassifier(n_estimators=100, random_state=42)
    clf.fit(X_train_scaled, y_train)
    y_pred = clf.predict(X_test_scaled)
    y_proba = clf.predict_proba(X_test_scaled)[:, 1]
    print("📊 Classification Report:")
    print(classification_report(y_test, y_pred))
    print("🎯 ROC AUC Score:", roc_auc_score(y_test, y_proba))
    # Confusion Matrix
    sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt='d', cmap='Blues')
    plt.title("Confusion Matrix")
    plt.show()
    df_test = X_test.copy()
    df_test['actual'] = y_test
    df_test['churn_probability'] = y_proba
    df_test['high_risk'] = (df_test['churn_probability'] > 0.7).astype(int)
    # Top 10 likely churners
    print("🔎 Top High-Risk Passengers:")
    return display(df_test[df_test['high_risk'] == 1].sort_values(by='churn_probability', ascending=False).head(10))   

if __name__ == "__main__":
    print("dataframe: ")
    display(df)
    print("Customer Type Distribution :")
    countplot(df,'customer type')
    high_value_low_satisfaction = high_value_low_satisfaction(df)
    print("Class andT Type of Travel Distribution of High_value_low_satisfaction Passengers :")
    countplot(high_value_low_satisfaction,'class','type of travel')
    segment_summary_correlation(high_value_low_satisfaction)
    correlation_df = Inflight_features_correlation(df)
    ttest_df = ttest_results(df)
    print("Inflight Services Correlation Map of High_value_low_satisfaction Passengers :")
    barplot(correlation_df,'Correlation')
    recommendations(ttest_df,correlation_df)
    delay_satisfaction(df)
    randomForestModelBuilding(df)

```
